```python
import numpy as np
from matplotlib import pyplot as plt
from utils.make_dfu import make_dfu
```

```python
dfu = make_dfu()
```

```python
I,J = (8,8)
fig, axs = plt.subplots(I,J, figsize=(15, 15))

for i in range(I):
    for j in range(J):
        l = np.random.choice(dfu.script.unique())
        D = dfu[dfu.script==l].sample(1).iloc[0]
        a = axs[i,j]
        a.imshow(D.picture,cmap='gray')
        a.axis(False)
        a.set_title(D.script)
```
